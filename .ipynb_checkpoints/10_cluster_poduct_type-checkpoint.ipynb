```python
import pywt
import pandas as pd
import numpy as np
import seaborn as sns
import datetime
from functools import reduce
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
product = pd.read_pickle('./datasets/train_product_series.pkl')
```

# CLUSTER PRODUCT INTO TYPES
--------------

## K-Means

```python
a = product.sort_values('date').groupby('product_id')['sessions'].apply(lambda x: list(x)).values
```

```python
a = np.array(list(a))
```

```python
result = []
for k in range(2,30):
    cluster = KMeans(n_clusters=k, max_iter=300)
    cluster.fit(a)
    result.append(cluster.inertia_)
plt.plot(result)
```

```python
cluster = KMeans(n_clusters=3, max_iter=300)
cluster.fit(a)
labels = pd.DataFrame(cluster.labels_, columns = ['product_labels'])
product_label = pd.merge(left = product, right = labels, left_on='product_id', right_index=True)
product_label.head().T
```

```python
sns.relplot(data = product_label, x = 'date', y = 'sessions', col = 'product_labels', hue = 'product_labels', kind = 'line')
```

## K-iMeans

* Compare product time series
  * Decompose the serie to reduce dimensionality
  * Start with a low value decomposition
* Iterate over K values
  * Along with iterations, iterate through decomposition dimensionality
  * initiate the centroids of each iteration with the (k-1)th dimension projection projected in the kth dimension

----------
### Haar decomposition

#### Requirements
To be able to use haar decomposition we need to have time series of the same length

```python
product.groupby('product_id')['date'].apply(lambda x: len(np.unique(x))).hist()
```

```python
product.groupby('product_id')['date','bookings','sessions'].apply(lambda x: np.sum(x.isna())).hist()
```

So it seems that we all the products are complete time series of the same length

----------
#### Decomposition
Example of haar decomposition on the highest level,ie time serie length /2

```python
test = pywt.dwt(product[product.product_id == 89].sort_values('date').sessions.tolist(),'haar')
decomp = [(datetime.datetime(2013,1,1) + datetime.timedelta(days = i),j) for i,j in enumerate(reduce(lambda x,y: x+y,[[i,i] for i in test[0]]))]
dtest = pd.DataFrame(decomp,columns = ['date','sessions'])
dtest['type'] = 'decomp'
dt = product[product.product_id == 89].loc[:,['date','sessions']]
dt['type'] = 'original'
dt_plot = pd.concat([dt,dtest],axis = 0)
sns.relplot(data=dt_plot, x = 'date',y='sessions',hue = 'type',kind='line', aspect= 4)
```

Example of Haar decomposition on lower levels

```python
testmulti = pywt.wavedec(product[product.product_id == 89].sort_values('date').sessions.tolist(),'haar')
#pywt.waverec(testmulti,'haar').tolist()[:5]
testmulti
```

```python
for i in range(0,len(testmulti)-1):
    testmulti[i] = np.zeros_like(testmulti[i])
pywt.waverec(testmulti,'haar')
```
